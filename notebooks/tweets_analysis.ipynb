```python
#IMPORTS

import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
```

```python
data = pd.read_csv('../data/dataset_twitter-scraper_2023-02-16_09-48-43-259.csv')
```

```python
data.head(5)
```

```python
data.dropna(axis=1, how='all')
```

```python
data.head(5)
```

```python
data = data.drop_duplicates(subset='full_text')
data = data[~data['full_text'].str.startswith('RT ')]
data = data.drop(columns=['id', 'conversation_id', 'created_at', 'hashtags/0', 'hashtags/1', 'hashtags/2', 'hashtags/3', 'hashtags/4','hashtags/5' ,'hashtags/6' ,'hashtags/7' ])
```

```python
data.head(5)
```

```python
# Tokenize the text

data['tokens'] = data['full_text'].apply(word_tokenize)
```

```python
# Remove stop words

stop_words = set(stopwords.words('english'))
data['tokens'] = data['tokens'].apply(lambda x: [word for word in x if word not in stop_words])
```

```python
# Stem the words

stemmer = SnowballStemmer('english')
data['tokens'] = data['tokens'].apply(lambda x: [stemmer.stem(word) for word in x])
```
